```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_movies= pd.read_csv('PI_MLOps/movies_dataset.csv')
data_credits= pd.read_csv('PI_MLOps/credits.csv')
```

En esta parte hacemos todos los puntos que se piden menos el desanidado de las columnas 

```python
#Eliminamos las columnas que no se van a utilizar: video,imdb_id,adult,original_title,poster_path y homepage.
data_movies= data_movies.drop(['video','imdb_id','adult','original_title','poster_path','homepage'], axis=1)
#Los valores nulos del campo release date deben eliminarse.
data_movies= data_movies.dropna(subset=['release_date'])
#Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.
values = {"revenue": 0, "budget": 0}
data_movies=data_movies.fillna(value=values)
#Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo 
# estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.
data_movies['revenue']=pd.to_numeric(data_movies['revenue'], errors ='coerce')
data_movies['budget']=pd.to_numeric(data_movies['budget'], errors ='coerce')
data_movies['return']=data_movies['revenue'].div(data_movies['budget']).replace(np.inf, 0).replace(np.NaN,0)
#De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.
len('1995-10-30')
from datetime import datetime, date

# Definimos la función para convertir un string de fecha a un objeto date
def convert_to_date(date_string):
    if len(date_string)!=10:
        date_string= np.nan
    else:
        return datetime.strptime(date_string, '%Y-%m-%d')


#Aplicamos la función a la columna release_date
data_movies['release_date']= data_movies['release_date'].apply(convert_to_date)

data_movies['release_date'].isnull().sum()
data_movies= data_movies.dropna(subset=['release_date'])
#Creamos la columna con sólo el año 
data_movies['release_year']=data_movies['release_date'].dt.year
```

En esta parte se hace el desanidado de las columnas 

```python
#convertirmos la columna id de objeto a entero para poder hacer el merge
data_movies['id'] = data_movies['id'].astype(int)
#Juntamos los dataframes:créditos y las películas
data_movies_credits=data_movies.merge(data_credits,on='id')


#Vericamos que no haya valores repetidos de 2 formas:
data_movies_credits['id'].value_counts()
data_movies_credits.duplicated().sum()
#Y los eliminamos
data_movies_credits.drop_duplicates(inplace=True)
```

```python
#Creamos función que convierte los datos de string a:  diccionario o lista de diccionario
import ast 
def convert_to_dicc(string_a_dicc):
    if type(string_a_dicc)== str:
        return ast.literal_eval(string_a_dicc)
    else:
        return  np.nan

#Aplicamos la función a las columnas 
data_movies_credits['genres']= data_movies_credits['genres'].apply(convert_to_dicc) #lista de diccionario
data_movies_credits['belongs_to_collection']=data_movies_credits['belongs_to_collection'].apply(convert_to_dicc) #diccionario
data_movies_credits['production_companies']=data_movies_credits['production_companies'].apply(convert_to_dicc) #lista de diccionario
data_movies_credits['production_countries']=data_movies_credits['production_countries'].apply(convert_to_dicc) #lista de diccionario
data_movies_credits['spoken_languages']=data_movies_credits['spoken_languages'].apply(convert_to_dicc) #lista de diccionario
data_movies_credits['cast']=data_movies_credits['cast'].apply(convert_to_dicc) #lista de diccionario
data_movies_credits['crew']=data_movies_credits['crew'].apply(convert_to_dicc) #lista de diccionario
```

```python
#Creamos función que regresa en forma de LISTAS los campos keys de nombre
#Funciona para las columnas: genre, companies, countries, languages, cast
def convertir_a_lista(string_a_dicc):
    if type(string_a_dicc)== list:
        lista_vacia=[]
        for i in string_a_dicc:
            lista_vacia.append(i['name'])
        return lista_vacia
    else:
        return  np.nan
    

#Creamos función que regresa en forma de STRING  los campos keys= director
#Funciona para las columna: crew
def convertir_a_director(string_a_dicc):
    if type(string_a_dicc)== list:
        for i in string_a_dicc:
            if i['job']=='Director': 
                return i['name']
    else:
        return  np.nan
    #Creamos función que regresa en forma de LISTA los campos keys= director
#Funciona para las columna: crew
def convertir_a_director_lista(string_a_dicc):
    if type(string_a_dicc)== list:
        lista_director=[]
        for i in string_a_dicc:
            if i['job']=='Director': 
                lista_director.append(i['name'])
                return lista_director
    else:
        return  np.nan


#Creamos función que regresa listas con el nombre de la coleccion
#Funciona para las columna: belongs_to_collection
def lista_collection(string_a_dicc):
    if type(string_a_dicc)== dict:
        return string_a_dicc['name']
    else:
        return  np.nan
    
    
#Aplicamos la función para cada columna 
data_movies_credits['genres']= data_movies_credits['genres'].apply(convertir_a_lista)
data_movies_credits['production_companies']=data_movies_credits['production_companies'].apply(convertir_a_lista)
data_movies_credits['production_countries']=data_movies_credits['production_countries'].apply(convertir_a_lista)
data_movies_credits['spoken_languages']=data_movies_credits['spoken_languages'].apply(convertir_a_lista)
data_movies_credits['cast']=data_movies_credits['cast'].apply(convertir_a_lista)
data_movies_credits['crew']=data_movies_credits['crew'].apply(convertir_a_director_lista)
data_movies_credits['belongs_to_collection']=data_movies_credits['belongs_to_collection'].apply(lista_collection)
```

A partir de aqui se crea otro dataframe llamado "data_modelo" para Machine Learning. 

```python
#Selección de columnas y cambio de nombre a dichas columnas 
data_modelo=data_movies_credits[['id','title','original_language','overview','genres','cast','crew']]
data_modelo.columns = ['id_pelicula', 'titulo','idioma_original', 'sinopsis', 'genero', 'actores', 'director']
```

```python
#La columna sinopsis la convertimos en lista
def quitar_espacios_sinopsis(texto):
    if type(texto) == str:
        #print('Es texto')
        return texto.split()
    else:
        return np.nan 
    
#Aplicamos la función
data_modelo['sinopsis']= data_modelo['sinopsis'].apply(quitar_espacios_sinopsis)
```

```python
#Creamos una copia del modelo 
data_modelo_copia=data_modelo
```

```python
#Quitamos espacios de las columnas GENERO, ACTORES y Director

#Funciona para la columna GENERO y ACTORES, quita espacios de las i cuando es tipo [(i),(i),(i),(i)]
data_modelo_copia['genero']=data_modelo_copia['genero'].apply(lambda x:[i.replace(" ","") for i in x])
data_modelo_copia['actores']=data_modelo_copia['actores'].apply(lambda x:[i.replace(" ","") for i in x])

#Funciona para la columna DIRECTOR
#Funciona si es tipo lista
def quitar_espacios_director(texto):
    if type(texto) == list:
        lista_vacia=[]
        for i in texto: 
            lista_vacia.append(i.replace(" ",""))
        return lista_vacia
    else:
        return np.nan 

#Se aplica la función a la columna DIRECTOR 
data_modelo_copia['director']= data_modelo_copia['director'].apply(quitar_espacios_director)
```

```python
#ESTE NO SIRVIÓ PORQUE ME MARCABA TAGS= NAN

#Se crea la columna tags con toda la información de las demás columnas, 
#Esto sólo se puede hacer si todas las columnas que estamos sumando son de tipo lista


#data_modelo_copia['tags']=data_modelo_copia['sinopsis']+data_modelo_copia['actores']+ data_modelo_copia['genero']+data_modelo_copia['director']

```

```python
#Se crea la columna tags con toda la información de las demás columnas
data_modelo_copia['tags'] = (
    data_modelo_copia['titulo'].fillna('') + ' ' +
    data_modelo_copia['idioma_original'].fillna('') + ' ' +
    data_modelo_copia['sinopsis'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '').replace(np.nan, '') + ' ' +
    data_modelo_copia['genero'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '').replace(np.nan, '') + ' ' +
    data_modelo_copia['actores'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '').replace(np.nan, '') + ' ' +
    data_modelo_copia['director'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '').replace(np.nan, '')
)

#Quito las mayúsculas de la columna tags
data_modelo_copia['tags'] = data_modelo_copia['tags'].str.lower()
```

```python
#Ahora que se ha creado la columna tags, 
# Se re seleccionan las columnas para simplificar el dataframe 
data_tags=data_modelo_copia[['id_pelicula', 'titulo','tags']]
data_tags
```

Hay muchas palabras que tienen la misma raíz:
De esto: ['loved', 'loving', 'love']
Necesitamos que sea así: ['love','love','love']

```python
import nltk 
from nltk.stem.porter import PorterStemmer

#Creamos el objeto 
ps=PorterStemmer()

def stem(texto):
    y=[]
    #Aquí se itera a través de las palabras en el texto, 
    # usando el método split() que divide el texto en palabras separadas 
    # por espacios en blanco. En cada iteración, una palabra se 
    # almacena temporalmente en la variable i.
    for i in texto.split():
        #La función stem() se aplica a cada palabra i,
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#La función se aplica para el primer valor para saber que funciona 
stem(data_tags['tags'][0])
```

```python
data_tags['tags']=data_tags['tags'].apply(stem)
```

```python
#CountVectorizer es una herramienta de procesamiento de lenguaje natural (NLP) que se utiliza para 
#convertir una colección de documentos de texto en una matriz de recuento de palabras o tokens.
from sklearn.feature_extraction.text import CountVectorizer
#En este caso, se seleccionan las 2500 palabras más frecuentes 
#Este parámetro especifica que las palabras de parada (stop words) en inglés deben ser excluidas del análisis. 
#Las palabras de parada son palabras comunes que a menudo no aportan un significado relevante en el análisis de texto
#(por ejemplo: "el", "es", "y", "un", etc.). Al especificar 'english', CountVectorizer utilizará una lista
#predeterminada de palabras de parada en inglés.
cv = CountVectorizer(max_features=2500,stop_words='english')
```

```python
#Se aplica el CountVectorizer a la columna tags
#Es decir, se crearon vectores para cada película 
vectors= cv.fit_transform(data_tags['tags']).toarray()
```

```python
#Número de películas que tenemos en nuestro dataframe: 29681
#LAs 5000 palabras que especificamos 
print('El arreglo de vectores tiene una dimen de ',vectors.shape ) #salida --> (44936, 2500)

#Este sería la matriz para la primer película
print(vectors[0])

## Te devuelve el array con los nombres de las películas
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#Se crea la matriz de similarididad, es decir, ya que hemos creado un vector por cada película
#Se mide la distancia entre ellos para deducir la similitud. Entre más distancia menos similitud
#distancia = 1/similitud   (inversamente proporcional )
#Para 1 vector (1 película) se calculan 44936  ángulos, es decir una matriz de  (44936X44936)
similarity = cosine_similarity(vectors)
```

```python
def recomendacion(pelicula):
    #Busca el índice de la película que queremos
    movie_index= data_tags[data_tags['titulo']==pelicula].index[0]
    distancia =similarity[movie_index]
    # list(enumerate(similarity[0]))
    #Enumera cada elemento del array similarity[0] y lo guarda en 
    # forma de lista de tuplas es decir:
    # [(0, 1),(1, 0.11572751247156893),(2, 0.08559209850218259),,...]

    #Con enumerate lo Ordenamos de mayor a menor , 
    # key es para una condición al ordenar, en este caso
    #de la lista de tuplas la vas a ordenar tomando el valor x[1] (x,[0],x[1])
    #y me vas a mostrar los 5 primeros valores 
    #Así ya optenemos el índice de las películas que tienen más similitud
    top_5= sorted(list(enumerate(distancia)),reverse=True,key=lambda x:x[1])[1:6]
    recomendaciones=[]
    for j in top_5:
        recomendaciones.append(data_tags.iloc[j[0]].titulo)
    return recomendaciones 
```

```python
#probamos que la función funcione correctamente
recomendacion('Father of the Bride Part II')
```

Aplicamos la función de recomendación para cada una de las películas que tenemos y creamos un nuevo dataframe para poder usarlo en el deploy y facilitarlo

```python
# Aplicar la función a la columna "pelicula_column"
data_tags['recomendaciones'] = data_tags['titulo'].apply(recomendacion)
```
